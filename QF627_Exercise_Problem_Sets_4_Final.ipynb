### <font color = red> Question 4</font>. Based on your lines of codes that lead to answer 3 above, please calculate the maximum drawdown and the longest drawdown period.

```python
stock_data_IBM["max_gross_performance"] =\
(
    stock_data_IBM
    ["IBM_strategy_cumulative_returns"]
    .cummax()
)
```

```python
stock_data_IBM['Drawdown'] =  stock_data_IBM["IBM_strategy_cumulative_returns"] / stock_data_IBM["max_gross_performance"] -1

stock_data_IBM['Drawdown']
```

```python
stock_data_IBM['MaxDrawdown'] = stock_data_IBM['Drawdown'].cummin()
```

```python
# Initialize 'HighWaterMarkDate' column with NaN or stock_data_IBM's index, which is likely the date
stock_data_IBM['HighWaterMarkDate'] = pd.NaT

# Iterate over the rows
for row in range(1, len(stock_data_IBM)):
    if stock_data_IBM['Drawdown'].iloc[row] == 0:
        # Set the HighWaterMarkDate as the current row's index (assumed to be date)
        stock_data_IBM.loc[stock_data_IBM.index[row], 'HighWaterMarkDate'] = stock_data_IBM.index[row]
    else:
        # Carry forward the previous row's HighWaterMarkDate
        stock_data_IBM.loc[stock_data_IBM.index[row], 'HighWaterMarkDate'] = stock_data_IBM['HighWaterMarkDate'].iloc[row - 1]

# Ensure the 'HighWaterMarkDate' column is in datetime format
stock_data_IBM['HighWaterMarkDate'] = pd.to_datetime(stock_data_IBM['HighWaterMarkDate'])
```

```python
#Calculate duration of drawdown
stock_data_IBM['DrawdownDuration'] = stock_data_IBM['HighWaterMarkDate']-stock_data_IBM.index
```

```python
#Calculate duration of maximum drawdown
stock_data_IBM['MaximumDrawdownDuration'] = stock_data_IBM['DrawdownDuration'].cummin()
```

```python
Maximum_drawdown = -100*stock_data_IBM.loc[stock_data_IBM.index[-1], 'MaxDrawdown']
Maximum_drawdown
```

```python
Longest_drawdown = -stock_data_IBM.loc[stock_data_IBM.index[-1], 'MaximumDrawdownDuration']
Longest_drawdown
```
